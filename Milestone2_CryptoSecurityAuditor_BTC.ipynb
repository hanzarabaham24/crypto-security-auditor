Dataset: Bitcoin (BTC)


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
plt.rcParams['figure.figsize'] = (14,5)

print("Environment ready ✅")
```

```

df = pd.read_csv('bitcoin.csv')
print(df.head())
```

```

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date']).sort_values('date')

df['market_cap'] = df['market_cap'].interpolate()
```

```

df['price_return_pct'] = df['price'].pct_change()*100
df['price_ma30'] = df['price'].rolling(30).mean()
df['volume_ma30'] = df['total_volume'].rolling(30).mean()

rolling_std = df['price'].rolling(30).std()
df['anomaly_score'] = (df['price'] - df['price_ma30']) / rolling_std

df['volume_spike'] = df['total_volume'] > (2*df['volume_ma30'])
df['rolling_volatility'] = df['price_return_pct'].rolling(30).std()

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```

# Chart 1
plt.plot(df['date'], df['price'])
plt.plot(df['date'], df['price_ma30'])
plt.title("Price + Moving Average")
plt.show()

# Chart 2
plt.plot(df['date'], df['total_volume'])
plt.title("Volume")
plt.show()

# Chart 3
plt.plot(df['date'], df['anomaly_score'])
plt.axhline(2); plt.axhline(-2)
plt.title("Anomaly Score")
plt.show()

# Chart 4
plt.plot(df['date'], df['market_cap'])
plt.title("Market Cap")
plt.show()

# Chart 5
df['price_return_pct'].hist()
plt.title("Returns Distribution")
plt.show()

# Chart 6
plt.scatter(df['date'], df['price'], s=1)
plt.title("Price Scatter")
plt.show()

# Chart 7
plt.scatter(df['total_volume'], df['price'], s=1)
plt.title("Price vs Volume")
plt.show()

# Chart 8
df.groupby('year')['price'].mean().plot(kind='bar')
plt.title("Yearly Avg Price")
plt.show()

# Chart 9
df['anomaly_score'].hist()
plt.title("Anomaly Distribution")
plt.show()

# Chart 10
pivot = df.groupby(['year','month'])['total_volume'].mean().unstack()
sns.heatmap(pivot)
plt.title("Heatmap Volume")
plt.show()

# Chart 11
sns.heatmap(df[['price','total_volume','market_cap']].corr(), annot=True)
plt.title("Correlation")
plt.show()

# Chart 12
top = df.nlargest(20,'anomaly_score')
plt.barh(top['date'].astype(str), top['anomaly_score'])
plt.title("Top Anomalies")
plt.show()

# Chart 13
plt.plot(df['date'], df['rolling_volatility'])
plt.title("Volatility")
plt.show()

# Chart 14
df.boxplot(column='price_return_pct', by='year')
plt.suptitle('')
plt.title("Returns by Year")
plt.show()

# Chart 15
plt.plot(df['date'], df['anomaly_score'])
plt.fill_between(df['date'], df['anomaly_score'], 0, where=df['anomaly_score']>2)
plt.fill_between(df['date'], df['anomaly_score'], 0, where=df['anomaly_score']<-2)
plt.title("Full Anomaly Timeline")
plt.show()
```


## 📌 Final Insights

- Anomaly scores highlight extreme Bitcoin events  
- Volume spikes correlate with price movements  
- Volatility increases during crashes  
- 2022 shows strongest negative anomalies  
- 2024 shows recovery and stabilization  

This supports the use of statistical monitoring for blockchain security signals.
